## Random Walk

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().strftime('%s'))
def getLinks(articleUrl):
    html = urlopen(f'http://en.wikipedia.org{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

visited_articles = set()
max_iterations = 50  # limit the loop to 50 articles

links = getLinks('/wiki/Kevin_Bacon')
iterations = 0

while links and iterations < max_iterations:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']

    # if article already visited, break the loop
    if newArticle in visited_articles:
        print(f"Already visited {newArticle}, stopping.")
        break

    # otherwise, add to visited and print
    visited_articles.add(newArticle)
    print(newArticle)

    # get new links from the selected article
    links = getLinks(newArticle)
    iterations += 1

print("Search completed or max iterations reached.")
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
import re
import time

# Set of visited pages to avoid revisiting
pages = set()

# Maximum number of links to follow
max_links = 100

def getLinks(pageUrl, depth=0, max_depth=3):
    if depth > max_depth or len(pages) >= max_links:  # Stop if depth is exceeded or max links reached
        return

    try:
        # Open the URL with a timeout to prevent hanging forever
        html = urlopen(f'http://en.wikipedia.org{pageUrl}', timeout=10)  # 10-second timeout
        bs = BeautifulSoup(html, 'html.parser')

        # Find all internal Wikipedia links
        for link in bs.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$')):
            if 'href' in link.attrs:
                newPage = link.attrs['href']

                # Only process new pages that haven't been visited
                if newPage not in pages:
                    print(newPage)  # Print the new link
                    pages.add(newPage)  # Add to set to avoid revisits

                    # Recursive call with incremented depth
                    getLinks(newPage, depth + 1, max_depth)

    except HTTPError as e:
        print(f"HTTP error retrieving page {pageUrl}: {e}")
    except URLError as e:
        print(f"URL error retrieving page {pageUrl}: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Start with a valid Wikipedia article, e.g., Kevin Bacon's page
start_time = time.time()
getLinks('/wiki/Kevin_Bacon')

print(f"Time taken: {time.time() - start_time} seconds")
```

## Collecting Data Across an Entire Site

```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
import re
import time

# set to store visited pages
pages = set()

# limit for the number of pages to visit
max_pages = 50

def getLinks(pageUrl, depth=0, max_depth=2, links_per_page=5):
    """
    Recursively retrieve links on Wikipedia pages, with limits on depth and links per page.
    """
    # stop recursion if depth exceeds max_depth or if we've already visited enough pages
    if depth > max_depth or len(pages) >= max_pages:
        return

    try:
        # open the URL with a timeout to prevent hanging forever
        html = urlopen(f'http://en.wikipedia.org{pageUrl}', timeout=10)
        bs = BeautifulSoup(html, 'html.parser')

        # print the page title and first paragraph
        print(bs.h1.get_text())

        bodyContent = bs.find('div', {'id':'bodyContent'}).find_all('p')
        if bodyContent:
            print(bodyContent[0].get_text())

        # print the edit link if available
        edit_link = bs.find(id='ca-edit')
        if edit_link:
            print(edit_link.find('a').attrs['href'])

    except HTTPError as e:
        print(f"HTTP Error: {e.code} while trying to retrieve page {pageUrl}")
    except URLError as e:
        print(f"URL Error: {e.reason} while trying to retrieve page {pageUrl}")
    except AttributeError:
        print('This page is missing something! Continuing.')

    # find all internal links, limiting to the first few links per page
    link_count = 0
    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']
            if newPage not in pages:
                print('-' * 20)
                print(newPage)
                pages.add(newPage)
                link_count += 1
                # limit the number of links followed per page
                if link_count > links_per_page:
                    break
                getLinks(newPage, depth + 1, max_depth, links_per_page)

                # shorter delay to avoid overwhelming the server but improve speed
                time.sleep(0.2)

# start with a valid Wikipedia article
getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
import re
import datetime
import random

# Set a seed for randomness to ensure results are reproducible
random.seed(datetime.datetime.now())

# Retrieves a list of all internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

# Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

# Retrieves a random external link
def getRandomExternalLink(startingPage):
    try:
        bs = BeautifulSoup(urlopen(startingPage), 'html.parser')
        externalLinks = getExternalLinks(bs, startingPage)
        if not externalLinks:
            print('No external links, looking around the site for one')
            internalLinks = getInternalLinks(bs, startingPage)
            if internalLinks:
                return getRandomExternalLink(random.choice(internalLinks))
            else:
                return None  # No internal links either
        else:
            return random.choice(externalLinks)
    except (HTTPError, URLError) as e:
        print(f"Error accessing {startingPage}: {e}")
        return None

# Follows external links recursively, with a limit on recursion depth
def followExternalOnly(startingSite, max_depth=5, current_depth=0):
    if current_depth >= max_depth:
        print("Reached max depth limit. Stopping.")
        return
    externalLink = getRandomExternalLink(startingSite)
    if externalLink:
        print(f'Random external link is: {externalLink}')
        followExternalOnly(externalLink, max_depth, current_depth + 1)
    else:
        print("No external links found to follow.")

# Start following external links from the given site
followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re

# Sets to track unique external and internal links
allExtLinks = set()
allIntLinks = set()

def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a', href=True):
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '' or parsed.netloc == netloc:
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
    return internalLinks

def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a', href=True):
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return externalLinks

def getAllExternalLinks(url, max_depth=2, depth=0):
    if depth > max_depth:
        return

    try:
        bs = BeautifulSoup(urlopen(url), 'html.parser')
    except (HTTPError, URLError) as e:
        print(f"Error accessing {url}: {e}")
        return

    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)

    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.add(link)
            print(f'External link: {link}')

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.add(link)
            # Recursively gather links, increasing depth by 1
            getAllExternalLinks(link, max_depth, depth + 1)

# Starting URL
start_url = 'https://www.oreilly.com'
allIntLinks.add(start_url)
getAllExternalLinks(start_url)
```
